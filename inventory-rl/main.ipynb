```python
from inventory_environment import InventoryEnv
from stable_baselines3 import DQN
```

```python
env = InventoryEnv()
```

### Treinar o agente usando DQN (Deep Q-Network)

```python
model = DQN("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=10000)
```

### Avaliar o modelo treinado

```python
obs, _ = env.reset()  # Gymnasium expects reset to return two values
for i in range(100):
    action, _states = model.predict(obs)
    obs, reward, done, truncated, info = env.step(
        action)  # step returns five values now
    env.render()
    if done:
        obs, _ = env.reset()
```
